```python
from selenium import webdriver
from selenium.webdriver.chrome.service import Service
from selenium.webdriver.common.by import By
from bs4 import BeautifulSoup
import pandas as pd
import time

# Set up Selenium WebDriver
options = webdriver.ChromeOptions()
options.add_argument("--headless")  # Optional: run in background
driver = webdriver.Chrome(service=Service(), options=options)

# Load the page
url = "https://www.the-numbers.com/market/"
driver.get(url)
time.sleep(3)  # Wait for content to load

soup = BeautifulSoup(driver.page_source, "html.parser")
driver.quit()

# Find the table
table = soup.find("table")
headers = [th.text.strip() for th in table.find_all("th")]

# Extract rows with original formatting
rows = []
for tr in table.find_all("tr")[1:]:
    cells = [td.text.strip() for td in tr.find_all("td")]
    if cells:
        rows.append(cells)

# Create DataFrame with original formatting
df = pd.DataFrame(rows, columns=headers)

# Display full values (no scientific notation)
pd.set_option("display.float_format", lambda x: f"${x:,.2f}")
display(df)
```

## Question 1

```python
import numpy as np
import matplotlib.pyplot as plt

def clean_percent(col):
    return (
        col.astype(str)                        # force everything to string
           .str.strip()                        # remove whitespace
           .str.replace('%','', regex=True)    # remove %
           .str.replace('<0.1','0.1', regex=False)  # replace "<0.1"
           .replace('', np.nan)                # handle empty strings
           .astype(float)                      # convert back to float
    )

df_merged['Domestic_percent'] = clean_percent(df_merged['Domestic_percent'])
df_merged['Foreign_percent']  = clean_percent(df_merged['Foreign_percent'])
```

### Display Graph

```python
domestic_foreign_trends = df_merged.groupby('year')[['Domestic_percent','Foreign_percent']].mean()

domestic_foreign_trends.plot(figsize=(10,6), marker='o')
plt.title("Balance of Domestic vs Foreign Box Office Revenue (2000–2024)")
plt.xlabel("Year")
plt.ylabel("Average % of Worldwide Revenue")
plt.grid(True)
plt.show()
```

```python
import numpy as np

def clean_currency(series):
    return (
        series.astype(str)                # ensure string
              .str.replace(',','', regex=False)   # remove commas
              .str.replace('$','', regex=False)   # remove dollar signs
              .str.replace('—','', regex=False)   # handle em-dash or missing values
              .replace('', np.nan)                # empty strings to NaN
              .astype(float)                      # convert to float
    )

# Apply to your revenue columns
df_merged['Worldwide'] = clean_currency(df_merged['Worldwide'])
df_merged['Domestic']  = clean_currency(df_merged['Domestic'])
df_merged['Foreign']   = clean_currency(df_merged['Foreign'])
```

### Display Graph

```python
# Calculate average worldwide revenue per film by year
avg_worldwide_per_film = df_merged.groupby('year')['Worldwide'].mean().sort_values(ascending=False)

# Display top 5 years
print("Top 5 Years by Average Worldwide Revenue per Film:")
print(avg_worldwide_per_film.head())

# Plot
import matplotlib.pyplot as plt
plt.figure(figsize=(10,6))
avg_worldwide_per_film.sort_index().plot(kind='bar', color='skyblue')
plt.title("Average Worldwide Box Office Revenue per Film (2000–2024)")
plt.xlabel("Year")
plt.ylabel("Average Worldwide Revenue ($)")
plt.xticks(rotation=45)
plt.tight_layout()
plt.show()
```

```python
import pandas as pd
import requests

url = "https://www.the-numbers.com/market/"
headers = {"User-Agent": "Mozilla/5.0"}

html = requests.get(url, headers=headers).text
tables = pd.read_html(html)

market_summary = tables[0]
ticket_prices = market_summary[['Year', 'Average Ticket Price']]
ticket_prices['Average Ticket Price'] = (
    ticket_prices['Average Ticket Price']
    .str.replace('$','', regex=False)
    .astype(float)
)

print(ticket_prices.head())
```

### Display Graph

```python
import matplotlib.pyplot as plt



plt.figure(figsize=(10,6))
plt.plot(ticket_prices['Year'], ticket_prices['Average Ticket Price'], 
         marker='o', color='darkred')

plt.title("Average Movie Ticket Prices Over Time (US)")
plt.xlabel("Year")
plt.ylabel("Average Ticket Price ($)")
plt.grid(True)
plt.tight_layout()
plt.show()
```

```python


# Sort by Year to ensure chronological order
ticket_prices = ticket_prices.sort_values('Year')

# Calculate year-over-year percent growth
ticket_prices['Percent Growth (%)'] = ticket_prices['Average Ticket Price'].pct_change() * 100

# Round for readability
ticket_prices['Percent Growth (%)'] = ticket_prices['Percent Growth (%)'].round(2)

# Display table
display(ticket_prices[['Year', 'Average Ticket Price', 'Percent Growth (%)']])
```

### Display Graph

```python
import matplotlib.pyplot as plt

ticket_prices = ticket_prices.sort_values('Year')
ticket_prices['Percent Growth (%)'] = ticket_prices['Average Ticket Price'].pct_change() * 100
ticket_prices['Percent Growth (%)'] = ticket_prices['Percent Growth (%)'].round(2)

# Plot bar chart
plt.figure(figsize=(10,6))
plt.bar(ticket_prices['Year'], ticket_prices['Percent Growth (%)'], color='steelblue')

plt.title("Year-over-Year Growth in Average Movie Ticket Prices (US)")
plt.xlabel("Year")
plt.ylabel("Percent Growth (%)")
plt.grid(axis='y', linestyle='--', alpha=0.7)
plt.xticks(rotation=45)
plt.tight_layout()
plt.show()
```
